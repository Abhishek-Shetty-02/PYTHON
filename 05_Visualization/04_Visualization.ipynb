```python
#Sales over a period of time

sales_by_date= df.groupby('Order Date')['Sales'].sum()
sales_by_date
```

```python
plt.figure(figsize=(20,5))
plt.plot(sales_by_date.index,sales_by_date.values, color='b',linestyle='--')
plt.title("Daily Sales Trend",fontsize=16)
# plt.xlable("Order date", fontsize=12)
# plt.ylable("Sales", fontsize=12)
# plt.xticks(rotation=90)
plt.xticks(rotation=45)
plt.grid(True, linestyle='--')
plt.show()
```

```python
#Daily Profit Trends

sales_by_date= df.groupby('Order Date')['Profit'].sum()
sales_by_date

plt.figure(figsize=(20,5))
plt.plot(sales_by_date.index,sales_by_date.values, color='b',linestyle='--')

plt.title("Daily Profit Trend",fontsize=16)
# plt.xlable("Order date", fontsize=12)
# plt.ylable("Sales", fontsize=12)
# plt.xticks(rotation=90)
plt.xticks(rotation=45)
plt.grid(True, linestyle='--')
plt.show()
```

```python
#Region Wise Sales

sales_by_region= df.groupby('Region')['Sales'].sum()
sales_by_region

plt.figure(figsize=(20,5))
plt.plot(sales_by_region.index,sales_by_region.values, color='b',linestyle='--')

plt.title("Region Wise Sales",fontsize=16)
# plt.xlable("Order date", fontsize=12)
# plt.ylable("Sales", fontsize=12)
# plt.xticks(rotation=90)
plt.xticks(rotation=45)
plt.grid(True, linestyle='--')
plt.show()
```

```python
#Region Wise Sales

sales_by_region= df.groupby('Region')['Sales'].sum().sort_values()
sales_by_region

plt.figure(figsize=(20,5))
plt.bar(sales_by_region.index,sales_by_region.values,color=['#2731F5','#F5275E','#CFF527','#27F5E4'])

# plt.bar(sales_by_region.index,sales_by_region.values,facecolor='w',edgecolor='b', color=['#2731F5','#F5275E','#CFF527','#27F5E4'])

plt.title("Region Wise Sales",fontsize=16)
# plt.xlable("Order date", fontsize=12)
# plt.ylable("Sales", fontsize=12)
# plt.xticks(rotation=90)
plt.xticks(rotation=45)
plt.grid(True, linestyle='--')
plt.show()
```

```python
#Region Wise Sales: Barhorizontal: Barh


sales_by_region= df.groupby('Region')['Sales'].sum().sort_values()
sales_by_region

plt.figure(figsize=(20,5))
plt.barh(sales_by_region.index,sales_by_region.values, color=['#2731F5','#F5275E','#CFF527','#27F5E4'])

plt.title("Region Wise Sales",fontsize=16)
# plt.xlable("Order date", fontsize=12)
# plt.ylable("Sales", fontsize=12)
# plt.xticks(rotation=90)
plt.xticks(rotation=45)
plt.grid(True, linestyle='--')
plt.show()
```

Based on the bar chart titled "Region Wise Sales", here are some key insights:

🔍 Insights:

Highest Sales – West Region:

The West region recorded the highest sales, slightly above 720,000 units (or currency).

This indicates strong market performance or higher customer demand in this region.

Second Highest – East Region:

The East region comes next, with sales close to 690,000.

This suggests it is also a strong-performing region, possibly due to effective sales strategies or high population density.

Moderate Sales – Central Region:

The Central region shows sales around 510,000.

While not the lowest, there's room for improvement compared to East and West.

Lowest Sales – South Region:

The South region has the lowest sales, around 390,000.

This may highlight challenges such as weaker market penetration, lower demand, or ineffective sales strategies.

📈 Recommendations:

Focus on the South Region: Investigate reasons behind low sales. Consider market research, promotions, or re-evaluating product-market fit.

Leverage Success in the West: Analyze what’s driving success here — sales tactics, customer demographics, or product fit — and replicate in weaker regions.

Growth Potential in Central: With moderate sales, targeted efforts could push this region into a higher-performing tier.

Monitor Trends: If this is time-based data, monitor over months/quarters to identify patterns or anomalies.
