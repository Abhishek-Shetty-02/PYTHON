```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# import seaborn
import plotly


df=pd.read_excel("EDUKRON_DATASETS/Source_Retail_Sales.xls")

df.head()
```

```python
#Line Graph

# Marker

plt.figure(figsize=(4,5))  #4-Width and 5-Heigth
plt.plot(df['Sales'].iloc[0:100].values,marker='o',markersize=5,markerfacecolor='white',markeredgecolor='black',markeredgewidth=1)  #marker
plt.xlabel("Index")
plt.ylabel("Sales")
plt.title("Index v/s sales")


# xlim and ylim: Cropping based on the dimensions

# plt.xlim(0,5)
# plt.ylim(0,100)
```

```python
#Line Graph

#plots
# Marker

plt.figure(figsize=(4,5))  #4-Width and 5-Heigth
plt.plot(df['Sales'].iloc[0:50].values,color='#1619DB')  #Hexcode
plt.plot(df['Sales'].iloc[50:100].values,color='#C43114')  

#.values on both plot: ndarray: converts both into [0:10] indexes
#Superimposition the graph
#helps to calualate D-D sales

plt.xlabel("Index")
plt.ylabel("Sales")
plt.title("Index v/s sales")
```

```python
#Subplot

plt.subplot(2,2,1)
plt.plot(df['Sales'].iloc[0:10])

plt.xlabel("Index")
plt.ylabel("Sales")
plt.title("Index v/s sales")

plt.subplot(2,2,2)
plt.plot(df['Sales'].iloc[10:20])
plt.xlabel("Index")
plt.ylabel("Sales")
plt.title("Index v/s sales")

plt.subplot(2,2,3)
plt.plot(df['Sales'].iloc[30:40])
plt.xlabel("Index")
plt.ylabel("Sales")
plt.title("Index v/s sales")

plt.subplot(2,2,4)
plt.plot(df['Sales'].iloc[40:50])
plt.xlabel("Index")
plt.ylabel("Sales")
plt.title("Index v/s sales")
```
